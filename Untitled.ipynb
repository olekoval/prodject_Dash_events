```python
code_names = {
    'C': 'Консультування та лікування',
    'L': 'Лабораторна діагностика',
    'I': 'Інструментальна діагностика',
    'P': 'Процедури',
    'E': 'Ургентні стани'
}

FILE_SERVICE = "./data/gr_services_first_letter_25_26.parquet"

def load_table_data():
    df = pl.read_parquet(FILE_SERVICE)
    df = df.with_columns(
        pl.col("first_letter_code")
        .replace_strict(code_names, default=None)
        .alias("name_service")
    )
    return df

def prepare_table_data():
    df = load_table_data()
    
    df_pivot = (
        df
        .select(["name_service", "year", "count_posluha"])
        .pivot(
            values="count_posluha",
            index="name_service",
            on="year"          # у нових версіях polars параметр "on" замість "columns"
        )
        .select(["name_service", "2025", "2026"])   # <-- явний порядок колонок
        .sort("name_service")
        .rename({"name_service": "Сервіси"})
    )
    return df_pivot

df = prepare_table_data()
df.head()
```

```python
code_names = {
        'C': 'Консультування та лікування',
        'L': 'Лабораторна діагностика',
        'I': 'Інструментальна діагностика',
        'P': 'Процедури',
        'E': 'Ургентні стани'
    }

def load_table_data():
    df = pl.read_parquet("./data/gr_services_Q_25_26.parquet")
    df = df.with_columns(
        pl.col("first_letter_code")
        .replace_strict(code_names, default=None)
        .alias("name_service")
    )
    return df
    
def service_q(Q):
    df_q = load_table_data()

    df_filter_q = (
        df_q
        .filter(pl.col("quarter") == Q)
        .select(["first_letter_code", "name_service", "year", "count_un_patient"])
        .pivot(
            values="count_un_patient",
            index=["first_letter_code", "name_service"],
            on="year"
        )
        .select(["first_letter_code", "name_service", "2025", "2026"])
        .sort("first_letter_code")
        .rename({"name_service": "Сервіси"})
        .with_columns(
            (pl.col("2026") - pl.col("2025"))
            .map_elements(
                lambda x: f"{x:+,}".replace(",", " "),
                return_dtype=pl.String,
            )
            .alias("Δ")
        )
        .with_columns(
            (
                (pl.col("2026") - pl.col("2025"))
                / pl.col("2025")
                * 100
            )
            .round(1)
            .map_elements(
                lambda x: f"{x:+.1f}%",
                return_dtype=pl.String,
            )
            .alias("Δ %")
        )
    )   # <-- цієї дужки не вистачало

    return df_filter_q

service_q("Q1")
```
